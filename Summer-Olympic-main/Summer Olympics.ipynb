# Summer Olympics Data Analysis 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("summer.csv")
```

### 1. In how many cities Summer Olympics is held so far?

```python
noCity=len(df['City'].unique())
print("Summer Olympics held in ",noCity," City")
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data = []

for Gold in df['Sport'].unique():
    data.append([Gold , len(df[df['Sport']  == Gold])])

pd.DataFrame(data,columns = ['Sport','Gold']).sort_values(by='Gold', ascending=False).head(1).plot(x = 'Sport', y = 'Gold', kind = 'bar', figsize = (5,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []

for Gold in df['Sport'].unique():
    data.append([Gold , len(df[df['Sport']  == Gold])])

pd.DataFrame(data,columns = ['Sport','Gold']).sort_values(by='Gold', ascending=False).head().plot(x = 'Sport', y = 'Gold', kind = 'bar', figsize = (5,5))
```

### 4. Which player has won most number of medals? (Top 5) 

```python
data = []

for medal in df['Athlete'].unique():
    data.append([medal , len(df[df['Athlete']  == medal])])

pd.DataFrame(data,columns = ['Athlete','medal']).sort_values(by='medal', ascending=False).head().plot(x = 'Athlete', y = 'medal', kind = 'bar', figsize = (5,5))
```
